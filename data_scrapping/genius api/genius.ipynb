```python
def search(artists_list, songs_of_art=5):
    all_songs = []
    for artist in artists_list:
        response = requests.get("https://api.genius.com/search",
            headers=headers,
            params={'q': artist, 'per_page': songs_of_art})

        if response.status_code == 200:
            hits = response.json()['response']['hits']

            for hit in hits:
                song = hit['result']
                detail_response = requests.get(f"https://api.genius.com/songs/{song['id']}", headers=headers)
                detail_json = detail_response.json() if detail_response.status_code == 200 else {}
                detail = detail_json.get('response', {}).get('song', {})

                album_data = detail.get('album') or {}
                album_name = album_data.get('name', '') if album_data else ''

                all_songs.append({
                    'id': song['id'],
                    'title': song['title'],
                    'artist': song['primary_artist']['name'],
                    'page_views': song.get('stats', {}).get('pageviews', 0),
                    'url': song['url'],
                    'release_date': detail.get('release_date', ''),
                    'album': album_name,
                    'label': detail.get('label', ''),
                    'language': detail.get('language', '')})

        time.sleep(0.3)

    return all_songs
```
