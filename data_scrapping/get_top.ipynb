### Топ 3000 треков скрампингом 

```python
import logging
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
URL = "https://kworb.net/spotify/artists.html"
```

```python
# Базовое логирование в stdout
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s %(levelname)s %(name)s %(message)s"
)
logger = logging.getLogger("kworb_scraper")
```

```python
def scrape_spotify_artists_selenium():
    logger.info("start: launching headless Chrome")
    opts = Options()
    opts.add_argument("--headless=new") # безинтерфейсный режим для скрапинга
    driver = webdriver.Chrome(options=opts)

    try:
        logger.info(f"navigate: {URL}")
        driver.get(URL)
        time.sleep(2)  # короткая пауза на рендер

        logger.info("locate: first <table>")
        table_el = driver.find_element(By.TAG_NAME, "table")
        html = table_el.get_attribute("outerHTML")

        logger.info("parse: reading HTML table into DataFrame")
        df = pd.read_html(html, flavor="bs4")[0]
        logger.info(f"done: rows={len(df)}, cols={len(df.columns)}")
        return df

    finally:
        logger.info("quit: closing browser")
        driver.quit()
```

```python
df = scrape_spotify_artists_selenium()
```
